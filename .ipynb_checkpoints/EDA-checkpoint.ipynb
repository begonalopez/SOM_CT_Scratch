# Exploratory Data Analisis

### 0. Load Data

```python
data_path = "scratch_projects.csv"
df = pd.read_csv(data_path)
```

```python
sns.set_palette("muted")
sns.set_style("whitegrid")  

# Create the bar plot for Mastery distribution
plt.figure(figsize=(10, 6))  # Set the figure size
sns.histplot(df['Mastery'], bins=21, kde=False, color='#4a4a4a')  

# Set plot title and labels
plt.title('Distribution of Mastery Scores', fontsize=12, weight='bold')
plt.xlabel('Mastery Score', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

sns.despine()

# Show the plot
plt.tight_layout()
plt.show()
```

### 1. Data Cleaning

Delete√á irrelevant features.

```python
# Delete project_ID and p_ID
df.drop(columns=['project_ID', 'p_ID', 'project-name', 'username'], inplace=True)
```

### 2. Statistical Analysis

```python
# Verify null values
df.isnull().sum()
```

We can observe that we doesn't have any nan value, insted of that we have -1 labels on the missing labels of social data (views, remixes, favorites, love).

```python
(df == -1).sum()
```

```python
# As there are few elements with missing values we can drop them
df = df.replace(-1, np.nan).dropna()
```

```python
# Set categorical features
categorical_columns = ['is-remix', 'Abstraction', 'Parallelism', 'Logic', 'Synchronization', 'FlowControl', 'UserInteractivity', 
                       'DataRepresentation', 'Mastery', 'InstancesSprites']

df[categorical_columns] = df[categorical_columns].astype('category')
```

```python
numeric_columns = df.select_dtypes(include=['number']).columns
```

```python
df.describe()
```

```python
df.describe(include='category')
```

### 3. Visual Analysis

```python
# Compute number of columns and rows
num_plots = len(numeric_columns)
num_cols = 3
num_rows = math.ceil(num_plots / num_cols)

fig, axs = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(20, num_rows * 4))
plt.subplots_adjust(hspace=0.5, wspace=0.3)

for column, ax in zip(numeric_columns, axs.ravel()):

    # Filter column and specify figure
    sns.histplot(df[column], kde=True, ax=ax)

    # Format
    ax.set_title(f"{column} distribution", fontsize=12)
    ax.set_xlabel(column, fontsize=10)
    ax.set_ylabel('Frequency', fontsize=10)
    ax.tick_params(axis='both', which='major', labelsize=8)

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(nrows=5, ncols=2, figsize=(30, 15))
plt.subplots_adjust(hspace=0.5)

for column, ax in zip(categorical_columns, axs.ravel()):

    # Compute distribution
    distribution_percentage = (df[column].value_counts(normalize=True) * 100).sort_index()

    sns.barplot(x=distribution_percentage.index, y=distribution_percentage.values, ax=ax)

    ax.set_title(f'{column} distribution')
    ax.set_xlabel(column)
    ax.set_ylabel('%')

    # Rotate labels
    ax.tick_params(axis='x', rotation=60, direction='out')
    
    # Add percentages to each bin
    for p in ax.patches:
      height = p.get_height()
      ax.text(p.get_x() + p.get_width() / 2., height + 0.5, f'{height:.2f}%', ha='center')

plt.tight_layout()
plt.show()
```

### 4. Correlation Matrix

```python
numeric_columns = df.select_dtypes(include=['float64', 'int64'])

correlation_matrix = numeric_columns.corr()

plt.figure(figsize=(20, 16))
sns.heatmap(correlation_matrix, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

# Mastery Level Cut

```python
# Count the number of zeros for each feature at each mastery level
zero_counts = df.groupby('Mastery').apply(lambda x: (x == 0).sum()).drop(columns='Mastery')

# Calculate the percentage of zeros for each feature at each mastery level
mastery_counts = df['Mastery'].value_counts().sort_index()
zero_percentages = zero_counts.div(mastery_counts, axis=0) * 100

# Prepare data for plotting
features = zero_percentages.columns  # Get feature names
mastery_levels = zero_percentages.index  # Get mastery levels

# Define the number of rows and columns for the subplots
num_rows = 3
num_cols = 5

fig, axes = plt.subplots(num_rows, num_cols, figsize=(30, 24), sharex=True, sharey=True)
axes = axes.flatten()  # Flatten the 2D array of axes for easy iteration

# Plot each feature in its own subplot
for i, feature in enumerate(features):
    ax = axes[i]
    bar_width = 0.35
    index = np.arange(len(mastery_levels))

    ax.bar(index, zero_percentages[feature], bar_width, label=feature)
    ax.set_title(f'{feature} zeros distribution')
    ax.set_xticks(index)
    ax.set_xticklabels(mastery_levels)
    ax.set_ylabel('%')
    
    # Rotate labels
    ax.tick_params(axis='x', rotation=60, direction='out')

fig.suptitle('Number of Zeros per Feature for Each Mastery Level', fontsize=16)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])  
plt.show()
```

```python
# Count the number of zeros for each feature at each mastery level
zero_counts = df.groupby('Mastery').apply(lambda x: (x == 0).sum()).drop(columns='Mastery').sum(axis=1).reset_index()
zero_counts.columns = ['Mastery', 'Count']

ax = zero_counts.plot.bar(x='Mastery', y='Count', rot=0)
plt.title('Number of Total Zeros for Each Mastery Level')

# Rotate labels
ax.tick_params(axis='x', rotation=60, direction='out')
```

```python
# Calculate the percentage of zeros in each column
zero_percentage = (df == 0).mean() * 100

# Sort the features by percentage of zeros
zero_percentage_sorted = zero_percentage.sort_values(ascending=False)

# Plotting the distribution of zeros across features
plt.figure(figsize=(12, 8))  # Set figure size

# Bar plot for zero percentage across features
sns.barplot(x=zero_percentage_sorted.values, y=zero_percentage_sorted.index, palette='Blues_d')

# Add title and labels
plt.title('Percentage of Zeros Across Features', fontsize=16, weight='bold')
plt.xlabel('Percentage of Zeros (%)', fontsize=12)
plt.ylabel('Features', fontsize=12)

# Show the plot
plt.tight_layout() 
plt.show()
```

```python
filtered_df = df[df['Mastery'].astype('int64')>=11]
```

```python
filtered_df
```

```python
# Calculate the percentage of zeros in each column for both df and filtered_df
zero_percentage_df = (df == 0).mean() * 100
zero_percentage_filtered_df = (filtered_df == 0).mean() * 100

# Combine both percentages into a DataFrame for easier plotting
zero_comparison = pd.DataFrame({
    'Original': zero_percentage_df,
    'Filtered': zero_percentage_filtered_df
}).sort_values(by='Original', ascending=False)  # Sort by original zeros percentage

# Plotting the comparison of zero percentages
plt.figure(figsize=(12, 8))  # Set figure size

# Bar plot with original and filtered zero percentages for each feature
zero_comparison.plot(kind='bar', figsize=(12, 8), color=['#4a4a4a', '#4682B4'])

# Add title and labels
plt.title('Comparison of Zeros Across Features: Original vs. Filtered', fontsize=16, weight='bold')
plt.xlabel('Features', fontsize=12)
plt.ylabel('Percentage of Zeros (%)', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()  # Adjust layout for better fit
plt.show()
```
